```python
import os
import yaml
file = open("C:\\Users\\g1238\\Documents\\tcs_hackaton\\data\\garage_doors\\IoT_normal_Garage_door_3.log","r")
lines = file.readlines()
def opener(path, flags):
    return os.open(path, flags, 0o777)
os.umask(0)
```
